```python
#Simulador do Algoritmo Earliest Deadline First para um conjunto de tarefas que chegam em tempo igual a zero

tempo_de_simulacao = 55
tempo_atual = 0
fila_de_prontos = []
processador_ocupado = False
#tarefa_i = [tempo_de_computacao, periodo, tempo_de_chegada, tempo_cumulativo_de_execucao, deadline, identificador]
#indice              0               1            2                       3                    4          5

#caso de teste 1:
#tarefa_0 = [20,100,0,0,0,0]
#tarefa_1 = [40,150,0,0,0,1]
#tarefa_2 = [100,350,0,0,0,2]
#tarefas = [tarefa_0, tarefa_1, tarefa_2] 

#caso de teste 2:
tarefa_0 = [10,20,0,0,20,0]
tarefa_1 = [25,50,0,0,50,1]
tarefas = [tarefa_0, tarefa_1]

tarefa_atual = []
while(tempo_atual <= tempo_de_simulacao):
    print("Tempo atual: " + str(tempo_atual)) #imprime tempo_atual
    if (processador_ocupado == True):
        tarefa_atual[3] += 1
        if tarefa_atual[3] != tarefa_atual[0]:
            print("Tarefa " + str(tarefa_atual[5]) + " executando...")
    for i in range(0,len(tarefas)):
        if tarefas[i][2] == tempo_atual:
            fila_de_prontos.append(tarefas[i])
    def sort_key(tarefa):
        return tarefa[4]
    fila_de_prontos.sort(key=sort_key)
    for i in range (0,len(fila_de_prontos)):
        fila_de_prontos[i][4] = fila_de_prontos[i][2] + fila_de_prontos[i][1]       
    if processador_ocupado == True and (len(fila_de_prontos) != 0 and fila_de_prontos[0][4] < tarefa_atual[4]):
        print("Tarefa " + str(tarefa_atual[5]) + " preemptada.")
        fila_de_prontos.append(tarefa_atual)
        tarefa_atual = fila_de_prontos[0]
        del fila_de_prontos[0] #remove primeira tarefa da fila de prontos
        fila_de_prontos.sort(key=sort_key) #reordena fila de prontos
        print("Tarefa " + str(tarefa_atual[5]) + " executando...")
    if processador_ocupado == False:
        if len(fila_de_prontos) != 0:
            tarefa_atual = fila_de_prontos[0]
            fila_de_prontos.pop(0) #remove primeira tarefa da fila de prontos
            fila_de_prontos.sort(key=sort_key)
            processador_ocupado = True
            print("Tarefa " + str(tarefa_atual[5]) + " executando...")
    if len(tarefa_atual)!=0 and tarefa_atual[3] == tarefa_atual[0]:
        print("Tarefa " + str(tarefa_atual[5]) + " concluida." )
        tarefa_atual[3] = 0 
        processador_ocupado = False #processador fica desocupado
        for i in range(0,len(tarefas)):
            if tarefa_atual[5] == tarefas[i][5]:
                tarefas[i][2] = tarefa_atual[4]
                tarefas[i][4] = tarefas[i][2] + tarefas[i][1]
        tarefa_atual = []
        if len(fila_de_prontos) != 0:
            tarefa_atual = fila_de_prontos[0]
            fila_de_prontos.pop(0) #remove primeira tarefa da fila de prontos
            fila_de_prontos.sort(key=sort_key) #reordena fila de prontos
            processador_ocupado = True
            print ("Tarefa " + str(tarefa_atual[5]) + " executando...")
    for i in range (0,len(fila_de_prontos)):
        if (tempo_atual >= fila_de_prontos[i][4]):
            print("Tarefa " + str(fila_de_prontos[i][5]) + " perde deadline!")
    if len(tarefa_atual) != 0:
        if tempo_atual >= tarefa_atual[4]:
            print("Tarefa " + str(tarefa_atual[5]) + " perde deadline!")
    tempo_atual += 1
    print("Tarefa atual: " + str(tarefa_atual))
    print("Fila de prontos: " + str(fila_de_prontos))
    print("Lista de tarefas " + str(tarefas))
    print("Processador ocupado: " + str(processador_ocupado))
    print()
```
